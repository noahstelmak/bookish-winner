## Dia da semana com placares mais altos

Crie um gráfico que responda em qual dia da semana os placares foram mais altos. O conceito de "placar mais alto" é a soma dos gols feitos no jogo.

Como a quantidade de jogos não é a mesma em todos os dias da semana (finais de semana tendem a ter mais jogos que em dias da semana - o que resultaria em placares mais altos nestes dias), você deverá encontrar uma forma de equalizar esta diferença na quantidade de jogos.

A resposta deverá ser um gráfico que demonstre claramente o comparativo dos dias.

```python
df['total_goals'] = df['gf'] + df['ga']

goals_per_day = df.groupby('day')['total_goals'].sum()
games_per_day = df.groupby('day').size()
avg_goals_per_game = goals_per_day / games_per_day
avg_goals_per_game = avg_goals_per_game.reindex(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.figure(figsize=(10,6))
ax = avg_goals_per_game.plot(kind='bar', color='royalblue')
ax.set_xticklabels(['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom'], rotation=45)
plt.title('Média de Gols por Jogo por Dia da Semana', fontsize=14)
plt.xlabel('Dia da Semana', fontsize=12)
plt.ylabel('Média de Gols por Jogo', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```

## Formação Tática

Qual a formação tática mais comum para cada equipe de acordo com o resultado obtido (Ex: para vitórias, a formação mais comum é X, para empates Y e para Z em derrotas).

**Importante** responda de forma que seja possível fazer comparativos entre equipes.

Responda com um gráfico.

```python
df0 = df.groupby(['team', 'formation', 'result'])['result'].count().reset_index(name="count")
df_pivot = df0.pivot_table(index='team', columns='result', values='count', aggfunc='sum', fill_value=0)

teams = df0['team'].unique()
for team in teams:
    team_data = df0[df0['team'] == team]
    pivot_team_data = team_data.pivot(index='formation', columns='result', values='count').fillna(0)
    
    ax = pivot_team_data.plot(kind='bar', stacked=False, figsize=(8, 6), color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    
    for p in ax.patches:
        ax.annotate(int(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    plt.title(f'Formações Usadas por {team} para Cada Resultado', fontsize=14)
    plt.xlabel(None, fontsize=12)
    plt.ylabel('Número de Ocorrências', fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

## Posse de bola

Qual a média de posse de bola da equipe mandante quando ela perdeu o jogo?

Responda em forma de gráfico para poder comparar entre todas as equipes.

```python
df_losses = df[(df['venue'] == 'Home')]

mean_possession_loss = df_losses.groupby('team')['poss'].mean().reset_index()
mean_possession_loss = mean_possession_loss.sort_values(by='poss', ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(mean_possession_loss['team'], mean_possession_loss['poss'], color='skyblue')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height:.1f}',
        ha='center',
        va='bottom',
        fontsize=10,
        color='black')


plt.title('Média de Posse de Bola nas Derrotas para Cada Equipe Mandante', fontsize=16)
plt.xlabel(None, fontsize=12)
plt.ylabel('Média de Posse de Bola (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

## Classificação

Qual a classificação final do campeonato?

Responda em forma de tabela, ordenada do primeiro colocado ao último.
Você deverá apresentar o brasão do time (pode ser obtido online [aqui](https://logodetimes.com/premier-league/), mas automatize a busca, não salve o logo localmente), o nome, a quantidade de pontos e a posição final no campeonato.

**OBS** Vitória vale 3 pontos, empates 1 ponto e derrota 0 pontos.

```python
df0 = df.groupby(['team', 'result'])['result'].count().reset_index(name="count")
df0 = df0.pivot_table(index='team', columns='result', values='count')
df0 = df0.reset_index()
df0['Pontos'] = df0['W'] * 3 + df0['D'] * 1

df0 = df0.sort_values(by='Pontos', ascending=False).reset_index(drop=True)

def get_team_logo_url(team_name):
    return f"https://logodetimes.com/times/{team_name.lower().replace(' ', '-')}-football-club/{team_name.lower().replace(' ', '-')}-football-club-256.png"

df0['img'] = df0['team'].apply(get_team_logo_url)

def render_table(df):
    html = '<table>'
    html += '<tr><th>Posição</th><th>Brazão</th><th>Equipe</th><th>Pontos</th></tr>'
    
    for index, row in df0.iterrows():
        html += f"<tr><td>{index + 1}</td>"
        html += f"<td><img src='{row['img']}' width='30' height='30'></td>"
        html += f"<td>{row['team']}</td><td>{row['Pontos']}</td></tr>"
    
    html += '</table>'
    return HTML(html)

# Renderizar a tabela final com o brasão e classificação
render_table(df)
```

## Público

Crie uma visualização que apresente o público total de cada equipe, e também um detalhamento por jogos.

Apresentar os dados usando a técnica "DrillDown" será considerado um "plus". A visualização DrillDown permite que gráficos sejam detalhados com cliques em itens do gráfico.

Um exemplo com tabelas pode ser encontrado [aqui](https://dash.plotly.com/dash-ag-grid/enterprise-master-detail).

Um exemplo usando gráficos pode ser encontrado [aqui](https://community.plotly.com/t/show-and-tell-drill-down-functionality-in-dash-using-callback-context/54403?u=atharvakatre)

```python
total_attendance = df.groupby('team')['attendance'].sum().reset_index()

output = widgets.Output()

def plot_total_attendance():
    plt.figure(figsize=(8, 6))
    plt.bar(total_attendance['team'], total_attendance['attendance'], color='lightblue')
    plt.title('Público Total por Equipe')
    plt.xlabel('Equipe')
    plt.ylabel('Público Total')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def plot_team_games(team):
    team_data = df[df['team'] == team]
    
    plt.figure(figsize=(8, 6))
    plt.bar(team_data['date'], team_data['attendance'], color='orange')
    plt.title(f'Público dos Jogos do {team}')
    plt.xlabel('Jogo')
    plt.ylabel('Público')
    plt.xticks(rotation=90, ha='right')
    plt.tight_layout()
    plt.show()

team_selector = widgets.Dropdown(
    options=df['team'].unique(),
    description='Selecione a Equipe:',
    disabled=False,
)

def on_team_selected(change):
    with output:
        output.clear_output()
        plot_team_games(change['new'])

team_selector.observe(on_team_selected, names='value')

plot_total_attendance()
display(team_selector, output)
```
